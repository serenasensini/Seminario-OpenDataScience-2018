```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
df = pd.DataFrame(np.random.randn(30,5))
print(df)
```

```python
for i in range(0,4):
    if i % 2 == 0:
        df.at[i*2, i] = "NaN"
    elif i % 3 == 0:
        df.at[i*3, i] = "NaN"
    elif i % 5 == 0:
        df.at[i*5, i] = "NaN"
```

```python
print(df)
```

```python
df.to_csv("df_missing_values.csv")
```

```python
df_missing = pd.read_csv("df_missing_values.csv")
```

```python
df_missing
```

```python
# elimina le righe contententi valori NaN
df_missing.dropna()
```

```python
# i valori NaN vengono rimpiazzati dal valore medio della colonna
df_missing.fillna(df_missing.mean())
```

```python
# i valori NaN vengono rimpiazzati dal valore della moda della colonna
df_missing.fillna(df_missing.mode())
```

```python
# uso del pacchetto scikit-learn
from sklearn.preprocessing import Imputer
help(Imputer)
```

```python
# uso un imputatore
imp = Imputer(missing_values = "NaN", strategy = "median", axis = 0)
# segnalo le colonne che contengono valori NaN
imp = imp.fit(df_missing.iloc[:, 0:3])
# sostituisco i valori del dataset con quelli a cui ho applicato la mediana tramite l'imputatore
df_missing.iloc[:, 1:4] = imp.transform(df_missing.iloc[:, 1:4])
```

```python
def encode_gender(gender):
    if gender.lower() == "male" or gender.lower() == "m":
        return "M"
    elif gender.lower() == "f" or re.match("fem", gender.lower()):
        return "F"
```

```python
df_preprocess_missing['gender_dummy'] = df_preprocess_missing.gender.apply(encode_gender)
```

```python
def encode_age(age):
    if int(age) < 10:
        return int(age)*10
    else:
        return int(age)
```

```python
df_preprocess_missing['age'] = df_preprocess_missing.age.apply(encode_age)
```
