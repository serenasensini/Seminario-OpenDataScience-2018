```python
import pandas as pd
import numpy as np
```

```python
# Per generare dati casuali, uso la funzione randn (numero_casi, numero_variabili)
df = pd.DataFrame(np.random.randn(10,5))
print(df)
```

```python
# Per ottenere lo stesso risultato, si usa la funzione seed della libreria numpy
np.random.seed(99)
pd.DataFrame(np.random.randn(10,5))
```
